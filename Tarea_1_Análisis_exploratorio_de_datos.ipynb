```python
diamonds["log_price"]=np.log1p(diamonds["price"])
diamonds["log_carat"]=np.log1p(diamonds["carat"])

sns.regplot(data=diamonds, x="log_carat", y="log_price")
plt.xlabel("Peso del diamante")
plt.ylabel("Precio del diamante")
plt.title("Relaci√≥n peso vs. precio de los diamantes")
plt.ylim(top=10.5)
plt.show() #Gr√°fico con valores modificados en escala logar√≠tmica
```
